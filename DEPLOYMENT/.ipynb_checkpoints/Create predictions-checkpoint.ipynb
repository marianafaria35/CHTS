```python
import pickle
filename = "Model_GB_INT.sav"

loaded_model = pickle.load(open(filename, 'rb'))

X=dataset_predict.replace(np.nan,-1).drop(['urg_episodio', 'False'], axis=1)

episodio=dataset_predict['urg_episodio']

valor_prev=loaded_model.predict(X)
predictions = loaded_model.predict_proba(X)

prob=[]

for d in predictions:
    maior_prob=0
    for y in d: 
        if y>maior_prob:
            maior_prob=y
    prob.append(maior_prob)

valor_prev=valor_prev[0:5]
prob=prob[0:5]    
    
l=0   
for x in valor_prev:
    u = episodio.iloc[l]
    cursor.execute ("""UPDATE OUTPUT_CLASSIFICACAO
                       SET ultimo=0
                       WHERE urg_episodio=%s and target='internamento'""" %(int(u)))
    rows = [(int(u),'internamento',type(loaded_model).__name__,int(valor_prev[l]),round(prob[l],4),1)]
    cursor.executemany("insert into OUTPUT_CLASSIFICACAO (urg_episodio,target,técnica,valor_previsto,grau_confianca,ultimo) values(:1,:2,:3,:4,:5,:6)", rows)
    l=l+1
connection.commit()
```
